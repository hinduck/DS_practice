```python
#import required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#data from ohio state, loading cause of death in the state from year 2012

#cause of deaths
cause = 'Chronic Diseases', 'Unintentional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others'

#percentile
percentile = [62, 5, 4, 2, 1, 26]
```

```python
#set the pie chart plot properties

#set the figure style
plt.figure(figsize=(10, 10), facecolor='white')

#explode the largest pie in the data set
explode = (0.05, 0, 0, 0, 0, 0)

#set pie chart properties
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)

#set axis equal to draw pe as circle
plt.axis('equal')

#set title of the pie chart
plt.title('Ohio State - 2012: Leading Causes of Death')

#show the plot (pie chart)
plt.show()
```
